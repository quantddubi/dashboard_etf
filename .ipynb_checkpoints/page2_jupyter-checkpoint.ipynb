```python
import streamlit as st
import pandas as pd
import plotly.graph_objects as go

# Streamlit 앱 제목
st.title("Stocks Fund Flow vs. Weekly Return")

# CSV 파일 불러오기
file_path = "page2_ff,aum.csv"  # 실제 파일 경로로 변경 필요
df_page2 = pd.read_csv(file_path, index_col=0)

# 주식 데이터 필터링
df_stocks = df_page2[df_page2.index == '주식']

df_stocks['대분류'] =['Country','Country','Country','Country','Country','Country','Country','Country','Country','Style','Style','Style','Style','Style','Sector','Sector','Sector','Sector','Sector','Sector','Sector','Sector','Sector','Sector']
df_stocks = df_stocks[['대분류','분류','ff/aum','return_1w']]

# 점도표 생성
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df_stocks['ff/aum'],
    y=df_stocks['return_1w'],
    mode='markers',
    marker=dict(size=10, color='blue', opacity=0.7),
    text=df_stocks['분류'],  # Hover text
))

# 레이아웃 설정
fig.update_layout(
    title="Stocks Fund Flow vs. Weekly Return",
    xaxis_title="Fund Flow / AUM",
    yaxis_title="Weekly Return (%)",
    template="plotly_white"
)

# Streamlit에서 그래프 표시
st.plotly_chart(fig)

# 데이터프레임도 화면에 표시
# st.write("Filtered Data (Stocks Only):")
# st.dataframe(df_stocks)
```
