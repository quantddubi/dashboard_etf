## 데이터 합치기

```python
universe_df = pd.read_csv('universe.csv')
universe_df = universe_df.set_index('Full_ticker')
ff_w_df = pd.read_csv('ff_w_recently.csv', index_col=0)
aum_w_df = pd.read_csv('aum_w_recently.csv', index_col=0)
```

```python
df = pd.concat([universe_df, ff_w_df, aum_w_df], axis= 1, join ="inner",ignore_index = True)
```

```python
df
```

```python
df = df[[0,9,8,2,3,4,5,6,7]]
df.columns = ['name','aum','ff','asset','Strategy1', 'Strategy2','Strategy3','DM/EM','Country']
```

```python
df['aum'] = df['aum'].fillna(0)
df['ff'] = df['ff'].fillna(0)
```

```python
df.to_csv('data.csv')
```

# 자산군 분류별 전처리

```python
# Fund Flow를 자산유형(asset)과 스타일1(c1) 기준으로 합산
fund_flow_summary1 = df.groupby(['asset', 'Strategy1'])[['ff','aum']].sum().reset_index()
fund_flow_summary2 = df.groupby(['asset', 'Strategy2'])[['ff','aum']].sum().reset_index()
fund_flow_summary3 = df.groupby(['asset', 'Strategy3'])[['ff','aum']].sum().reset_index()
fund_flow_summary4 = df.groupby(['asset', 'DM/EM'])[['ff','aum']].sum().reset_index()
fund_flow_summary5 = df.groupby(['asset', 'Country'])[['ff','aum']].sum().reset_index()
```

```python
fund_flow_summary1.columns = ['자산군','분류','Fund_Flow','AUM']
fund_flow_summary2.columns = ['자산군','분류','Fund_Flow','AUM']
fund_flow_summary3.columns = ['자산군','분류','Fund_Flow','AUM']
fund_flow_summary4.columns = ['자산군','분류','Fund_Flow','AUM']
fund_flow_summary5.columns = ['자산군','분류','Fund_Flow','AUM']
```

```python
fund_flow_df = pd.concat([fund_flow_summary1,fund_flow_summary2,fund_flow_summary3,fund_flow_summary4,fund_flow_summary5],axis=0)
fund_flow_df = fund_flow_df.round(1)
fund_flow_df
```

```python
asset_order = ['주식','채권','원자재','대체']
calssification_order =['DM',
                       'EM',
                       'US',
                       'Europe',
                       'Japan',
                       'China',
                       'Taiwan',
                       'Korea',
                       'india',
                       'Growth',
                       'Value',
                       'Large',
                       'mid',
                       'small',                
                       'Communication',
                       'discretionary',
                       'staples',
                       'energy',
                       'financial',
                       'healthcare',
                       'industrial',
                       'materials',
                       'technology',
                       'utilities',
                       'agg',
                       'bank_loan',
                       'convertible',
                       'corporate',
                       'government',
                       'inflation',
                       'mortage_back',
                       'A+',
                       'IG',
                       'HY',
                       'long',
                       'intermediate',
                       'short',
                       'diversified',
                       'Precious_metals',
                       'Energy&Fuels',
                       'industrial_metal',
                       'agriculture',
                       'multi_strategy',
                       'macro',
                       'equity_hedge']

fund_flow_df['자산군'] = pd.Categorical(fund_flow_df['자산군'], categories=asset_order, ordered=True)
fund_flow_df['분류'] = pd.Categorical(fund_flow_df['분류'], categories=calssification_order, ordered=True)

# # 정렬 수행
fund_flow_df_sorted = fund_flow_df.sort_values(by=['자산군','분류'])
```

```python
fund_flow_df_sorted
```

```python
fund_flow_df_sorted['ff/aum']= round(fund_flow_df_sorted['Fund_Flow']/fund_flow_df_sorted['AUM']*100,2)
fund_flow_df_sorted = fund_flow_df_sorted.dropna(subset=["분류"])
```

## 자산군별 수익률 데이터

```python
df_price = pd.DataFrame(df_raw['asset_price'].iloc[16:,])
df_price = df_price.rename(columns=df_raw['asset_price'].iloc[12,1:49])
```

```python
df_price.set_index('Unnamed: 0', inplace=True)
```

```python
return_1w = round(df_price.pct_change(periods=7)*100,1)
return_1m = round(df_price.pct_change(periods=30)*100,1)
```

```python
fund_flow_df_sorted['return_1w'] = fund_flow_df_sorted['분류'].map(return_1w.iloc[-1])
fund_flow_df_sorted['return_1m'] = fund_flow_df_sorted['분류'].map(return_1m.iloc[-1])
```

```python
new_order = ['자산군','분류','AUM','Fund_Flow','ff/aum','return_1w','return_1m']
fund_flow_df_sorted = fund_flow_df_sorted[new_order]
```

```python
fund_flow_df_sorted.to_csv('asset_ff,aum.csv',encoding='utf-8-sig', index=False)
```
