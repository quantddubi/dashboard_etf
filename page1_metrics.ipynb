```python
import pandas as pd
import yfinance as yf
# import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
```

```python
# 엑셀 파일 불러오기 (sheet_name으로 특정 시트 지정 가능)
df_raw = pd.read_excel("etf_raw_daily.xlsx", sheet_name=None)
```

```python
df_price = pd.DataFrame(df_raw['asset_price'].iloc[16:,])
df_price = df_price.rename(columns=df_raw['asset_price'].iloc[12,1:49])
df_price.set_index('Unnamed: 0', inplace=True)
```

```python
df_price
```

```python
asset_price = df_price[['US','Europe','Japan','China','Korea','Unnamed: 49','Unnamed: 50','Energy&Fuel','fx']]
```

```python
weekly_asset_price = asset_price[asset_price.index.weekday ==4]
weekly_asset_price
```

```python
stock_price_pct = pd.DataFrame([weekly_asset_price.iloc[:,:5].pct_change().iloc[-1]*100], index=None)
```

```python
stock_price_pct = stock_price_pct.reset_index(drop=True)
```

```python
bond_price_pct = pd.DataFrame([weekly_asset_price.iloc[:,5:7].iloc[-1] - weekly_asset_price.iloc[:,5:7].iloc[-2]])
```

```python
comm_price_pct = pd.DataFrame([weekly_asset_price['Energy&Fuel'].pct_change().iloc[-1]*100], columns = ['Energy&Fuel'])
```

```python
fx_price_pct = pd.DataFrame([weekly_asset_price['fx'].iloc[-1] - weekly_asset_price['fx'].iloc[-2]], columns =['fx']) 
```

```python
df = pd.concat([stock_price_pct, bond_price_pct, comm_price_pct, fx_price_pct],axis=1)
```

```python
df
```

```python
df.loc[len(df)] = weekly_asset_price.iloc[-1]
```

```python
df
```

```python
df.to_csv('./data/page1_metrics_data.csv')
```
