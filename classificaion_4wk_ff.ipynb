```python
import pandas as pd

# 엑셀 파일 불러오기 (sheet_name으로 특정 시트 지정 가능)
df_raw = pd.read_excel("etf_raw_daily.xlsx", sheet_name='FF_d')
```

```python
ff_d = pd.DataFrame(df_raw.iloc[13:,])
ff_d = ff_d.rename( columns = df_raw.iloc[10,1:])
ff_d = ff_d.set_index('Unnamed: 0')
ff_d
```

```python
ff_d_to_w = ff_d.resample('W-FRI').sum()
```

```python
ff_d_to_4wk_rolling_sum = ff_d_to_w.rolling(window=4).sum()
```

```python
ff_d_to_4wk_rolling_sum = ff_d_to_4wk_rolling_sum.iloc[3:]
```

```python
ff_d_to_4wk_rolling_sum
```

```python
universe_df = pd.read_csv('./data/universe.csv', index_col=0)
universe_df = universe_df.set_index('Full_ticker')
```

```python
etf_ff_4wk = pd.concat([universe_df,ff_d_to_4wk_rolling_sum.T.iloc[:,-36:]], axis=1,join ="inner")
```

```python
stock_ff_4wk_rolling = etf_ff_4wk[etf_ff_4wk['Asset']=="주식"].iloc[:,8:].sum()
bond_ff_4wk_rolling = etf_ff_4wk[etf_ff_4wk['Asset']=="채권"].iloc[:,8:].sum()
commodity_ff_4wk_rolling = etf_ff_4wk[etf_ff_4wk['Asset']=="원자재"].iloc[:,8:].sum()
```

```python
asset_ff_4wk_rolling = pd.concat([stock_ff_4wk_rolling,bond_ff_4wk_rolling,commodity_ff_4wk_rolling],axis=1)
```

```python
asset_ff_4wk_rolling.columns = ['Stock','Bond','Commodity']
```

```python
asset_ff_4wk_rolling = asset_ff_4wk_rolling.astype(int)
asset_ff_4wk_rolling
```

```python
asset_ff_4wk_rolling.to_csv('asset_ff_4wk_rolling.csv')
```

# aum 데이터

```python
df_AUM = pd.read_excel("etf_raw_daily.xlsx", sheet_name='AUM_d')
```

```python
aum_d = pd.DataFrame(df_AUM.iloc[13:,])
aum_d = aum_d.rename(columns = df_AUM.iloc[10,1:])
aum_d = aum_d.set_index('Unnamed: 0')

friday_df_aum = aum_d[aum_d.index.weekday == 4]
friday_df_aum = friday_df_aum.astype(float).round(1)
```

```python
etf_aum=pd.concat([universe_df,friday_df_aum.T.iloc[:,-36:]], axis=1,join ="inner")
```

```python
stock_aum_12w = etf_aum[etf_aum['Asset']=="주식"].iloc[:,8:].sum()
bond_aum_12w = etf_aum[etf_aum['Asset']=="채권"].iloc[:,8:].sum()
commodity_aum_12w = etf_aum[etf_aum['Asset']=="원자재"].iloc[:,8:].sum()
```

```python
asset_aum_12w = pd.concat([stock_aum_12w,bond_aum_12w,commodity_aum_12w],axis=1)
asset_aum_12w.columns = ['Stock','Bond','Commodity']
```

# 합치기

```python
asset_ff_aum_4wk_rolling_aum = asset_ff_4wk_rolling / asset_aum_12w *100
```

```python
asset_ff_aum_4wk_rolling_aum
```

```python
asset_ff_aum_4wk_rolling_aum = asset_ff_aum_4wk_rolling_aum.round(2)
```

```python
asset_ff_aum_4wk_rolling_aum
```

```python
asset_ff_aum_4wk_rolling_aum.to_csv('asset_ff_aum_4wk_rolling.csv')
```
