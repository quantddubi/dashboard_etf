```python
import streamlit as st
import pandas as pd
```

```python
file_path = "asset_ff,aum.csv"  # 업로드된 파일 경로에 맞게 수정
df = pd.read_csv(file_path)
```

```python
df
```

```python
# Streamlit 설정
st.set_page_config(page_title="ETF Fund Flow Dashboard", layout="wide")
st.title("📊 ETF Fund Flow & AUM Dashboard")

# 필터 추가
st.sidebar.header("Filters")
columns = df.columns.tolist()
selected_column = st.sidebar.selectbox("Select Column to Filter", columns)
unique_values = df[selected_column].unique()
selected_value = st.sidebar.selectbox(f"Filter by {selected_column}", unique_values)

# 필터 적용
df_filtered = df[df[selected_column] == selected_value]

# 데이터 테이블 표시
st.write("### Filtered Data Table")
st.dataframe(df_filtered)
```
