```python
import pandas as pd

# 엑셀 파일 불러오기 (sheet_name으로 특정 시트 지정 가능)
df_raw = pd.read_excel("etf_raw_daily.xlsx", sheet_name='FF_d')
```

```python
ff_d = pd.DataFrame(df_raw.iloc[13:,])
ff_d = ff_d.rename( columns = df_raw.iloc[10,1:])
ff_d = ff_d.set_index('Unnamed: 0')
ff_d
```

```python
ff_d_to_w = ff_d.resample('W-FRI').sum()
```

```python
universe_df = pd.read_csv('universe.csv', index_col=0)
universe_df = universe_df.set_index('Full_ticker')
```

```python
etf_ff = pd.concat([universe_df,ff_d_to_w.T.iloc[:,-36:]], axis=1,join ="inner")
etf_ff
```

```python
stock_ff_12w = etf_ff[etf_ff['Asset']=="주식"].iloc[:,8:].sum()
bond_ff_12w = etf_ff[etf_ff['Asset']=="채권"].iloc[:,8:].sum()
commodity_ff_12w = etf_ff[etf_ff['Asset']=="원자재"].iloc[:,8:].sum()
```

```python
asset_ff_12w = pd.concat([stock_ff_12w,bond_ff_12w,commodity_ff_12w],axis=1)
```

```python
asset_ff_12w.columns = ['Stock','Bond','Commodity']
```

```python
asset_ff_12w = asset_ff_12w.astype(int)
asset_ff_12w
```

```python
asset_ff_12w.to_csv('asset_ff_12w.csv')
```

# AUM_timeseries

```python
df_AUM = pd.read_excel("etf_raw_daily.xlsx", sheet_name='AUM_d')
```

```python
aum_d = pd.DataFrame(df_AUM.iloc[13:,])
aum_d = aum_d.rename(columns = df_AUM.iloc[10,1:])
aum_d = aum_d.set_index('Unnamed: 0')

friday_df_aum = aum_d[aum_d.index.weekday == 4]
friday_df_aum = friday_df_aum.astype(float).round(1)
```

```python
etf_aum=pd.concat([universe_df,friday_df_aum.T.iloc[:,-36:]], axis=1,join ="inner")
```

```python
stock_aum_12w = etf_aum[etf_aum['Asset']=="주식"].iloc[:,8:].sum()
bond_aum_12w = etf_aum[etf_aum['Asset']=="채권"].iloc[:,8:].sum()
commodity_aum_12w = etf_aum[etf_aum['Asset']=="원자재"].iloc[:,8:].sum()
```

```python
asset_aum_12w = pd.concat([stock_aum_12w,bond_aum_12w,commodity_aum_12w],axis=1)
asset_aum_12w.columns = ['Stock','Bond','Commodity']
```

```python
asset_aum_12w
```

# FF + AUM

```python
asset_ff_aum = asset_ff_12w / asset_aum_12w *100
```

```python
asset_ff_aum = asset_ff_aum.round(2)
```

```python
asset_ff_aum.to_csv('asset_ff_aum.csv')
```
